```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\acera\Downloads\WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Fix column names (important: remove spaces)
df.columns = df.columns.str.strip()

# List of columns (corrected names)
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplot grid (3 rows, 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

# Flatten axes for easy loop
axes = axes.flatten()

# Plot each countplot
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```
